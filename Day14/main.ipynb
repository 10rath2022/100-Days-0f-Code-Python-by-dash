```python
import random
import art
import game_data

# Function to format account data
def format_data(account):
    name = account['name']
    follower_count = account['follower_count']
    description = account['description']
    country = account['country']
    return f"{name}, a {description}, from {country}"

# Function to check the user's guess
def check_answer(user_guess, a_followers, b_followers):
    # Compare follower counts
    if a_followers > b_followers:
        return user_guess == "A"
    else:
        return user_guess == "B"

# Display ASCII art logo
print(art.logo)

# Initialize game variables
score = 0
game_should_continue = True
account_b = random.choice(game_data.data)

# Game loop
while game_should_continue:
    # Make account B the next account A
    account_a = account_b
    account_b = random.choice(game_data.data)

    # Ensure account A and B are not the same
    while account_a == account_b:
        account_b = random.choice(game_data.data)

    # Print the comparison details
    print(f"Compare A: {format_data(account_a)}")
    print(art.vs)
    print(f"Against B: {format_data(account_b)}")

    # Ask user for their guess
    guess = input("Who has more followers? Type 'A' or 'B': ").upper()

    print("\n"* 20)
    print(art.logo)
    
    # Get follower counts
    a_follower_count = account_a["follower_count"]
    b_follower_count = account_b["follower_count"]

    # Check if the user is correct
    is_correct = check_answer(guess, a_follower_count, b_follower_count)

    # Provide feedback and update score
    if is_correct:
        score += 1
        print(f"You're right! Current score: {score}.")
    else:
        print(f"Sorry, that's wrong. Final score: {score}.")
        game_should_continue = False
```
